## EXPLORATORY DATA ANALYSIS 
### (analyzing the datasets)

```python
import pandas as pd

# Load datasets
tutors = pd.read_csv("tutors.csv") # make sure to use the correct path location of your tutors file
students = pd.read_csv("students.csv") # make sure to use the correct path location of your students file

# Basic preview
print(" Tutors Dataset:")
display(tutors.head())

print(" Students Dataset:")
display(students.head())

# Basic info
print("\nTutors Info:")
print(tutors.info())

print("\nStudents Info:")
print(students.info())
```

```python
print("\nTutors Summary:")
print(tutors.describe())

print("\nStudents Summary:")
print(students.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,4))
sns.histplot(tutors["rating"], kde=True, bins=10)
plt.title("Tutor Rating Distribution ⭐")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()

#Checking where the most rated tutors belong in the dataset
```

```python
plt.figure(figsize=(10,4))
sns.scatterplot(data=tutors, x="experience_years", y="rating", hue="subject", alpha=0.8)
plt.title("Experience vs Rating (by Subject)")
plt.show()
```

```python
sns.lmplot(data=tutors, x="experience_years", y="rating", hue="subject")
plt.title("Experience vs Rating Trend (by Subject)")
```

#### Insight: Higher Experienced tutors usually get higher ratings as we can see in the scatter plot and trend regression line

```python
# Detecting outliers in hourly rates
plt.figure(figsize=(6,4))
sns.boxplot(data=tutors, x="hourly_rate")
plt.title("Tutor Hourly Rate Distribution 💰")
plt.show()
```

#### There seem to be no outliers here.

```python
tutor_subjects = tutors['subject'].value_counts().head(5)
tutor_subjects.plot(kind='bar', color='skyblue')
plt.title("Top 5 Tutor Subjects")
plt.ylabel("Number of Tutors")
plt.show()
```

### Insight: Among all the subjects , Science and Computer Science seem to be most taught by tutors

```python
plt.figure(figsize=(8,5))
sns.countplot(data=students, x="grade", hue="required_subject")
plt.title("Student Grades vs Subjects Needed 🎓")
plt.show()
```

### Insight: To see which grade students are requesting which subjects most.This can be helpful later for recommendation.

# Data Cleaning

##### Handling missing values

```python
print("Missing values in tutors:")
print(tutors.isnull().sum())
```

```python
print("\nMissing values in students:")
print(students.isnull().sum())
```

Insight: Fortunately, there no null values in dataset to be taken care of.

##### Handling duplicate values

```python
print("\nDuplicates:")
print("Tutors:", tutors.duplicated().sum())
print("Students:", students.duplicated().sum())
```

```python
#Drop duplicates if there are any
tutors.drop_duplicates(inplace=True)
students.drop_duplicates(inplace=True)
```
